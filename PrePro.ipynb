```python
import os
import cv2
import albumentations as A
from albumentations import HorizontalFlip, Rotate, RandomBrightnessContrast, GaussianBlur, Compose, OneOf

augmentation = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.Rotate(limit=15, p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
    A.GaussianBlur(blur_limit=3, p=0.2),
    A.GaussNoise(p=0.2)
])

def augment_and_save(image_path, output_subdir, count=5):

    image = cv2.imread(image_path)
    if image is None:
        print(f"Skipping {image_path} (not an image)")
        return
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_name = os.path.basename(image_path).split('.')[0]

    for i in range(count):
        augmented = augmentation(image=image)
        aug_image = augmented["image"]
        aug_image_path = os.path.join(output_subdir, f"{image_name}_aug_{i}.jpg")
        cv2.imwrite(aug_image_path, cv2.cvtColor(aug_image, cv2.COLOR_RGB2BGR))
        print(f"Augmented and saved: {aug_image_path}")
```

```python
def apply_augmentation(input_dir, output_dir, count=5):

    for root, dirs, files in os.walk(input_dir):
        relative_path = os.path.relpath(root, input_dir)
        output_subdir = os.path.join(output_dir, relative_path)
        os.makedirs(output_subdir, exist_ok=True)

        for file in files:
            image_path = os.path.join(root, file)
            augment_and_save(image_path, output_subdir, count=count)

input_dir = "/content/drive/MyDrive/Plant Dataset"
output_dir = "/content/drive/MyDrive/Augmented_Set"
apply_augmentation(input_dir, output_dir, count=5)
```
